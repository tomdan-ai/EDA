```python
import numpy as np

temps_array = np.array([22.5, 25.0, 19.8, 30.2, 28.1])

# 1. Number of temperature readings
num_readings = np.shape(temps_array)[0]
print('Number of temperature readings:', num_readings)

# 2. Average temperature
avg_temp = np.mean(temps_array)
print('Average temperature:', avg_temp)

# 3. Temperatures greater than 25 degrees
count_gt_25 = np.sum(temps_array > 25)
print('Temperatures greater than 25 degrees:', count_gt_25)

# 4. Maximum temperature
max_temp = np.max(temps_array)
print('Maximum temperature:', max_temp)

# 5. Temperatures ≥ 20 degrees
count_ge_20 = np.sum(temps_array >= 20)
print('Temperatures ≥ 20 degrees:', count_ge_20)

# 6. Standard deviation
std_temp = np.std(temps_array)
print('Standard deviation:', std_temp)

# 7. Minimum temperature
min_temp = np.min(temps_array)
print('Minimum temperature:', min_temp)

# 8. Temperatures equal to 25.0
count_eq_25 = np.sum(temps_array == 25.0)
print('Temperatures equal to 25.0:', count_eq_25)

# 9. Median temperature
median_temp = np.median(temps_array)
print('Median temperature:', median_temp)

# 10. Temperatures greater than 28 degrees
count_gt_28 = np.sum(temps_array > 28)
print('Temperatures greater than 28 degrees:', count_gt_28)

# 11. Sum of all temperatures
sum_temps = np.sum(temps_array)
print('Sum of all temperatures:', sum_temps)

# 12. Temperatures less than 22 degrees
count_lt_22 = np.sum(temps_array < 22)
print('Temperatures less than 22 degrees:', count_lt_22)

# 13. Average temperature after adding 2 degrees
avg_temp_plus_2 = np.mean(temps_array + 2)
print('Average temperature after adding 2 degrees:', avg_temp_plus_2)

# 14. Temperatures greater than 24 degrees (using np.where)
indices_gt_24 = np.where(temps_array > 24)[0]
count_gt_24 = len(indices_gt_24)
print('Temperatures greater than 24 degrees:', count_gt_24)

# 15. Variance of temperatures
var_temp = np.var(temps_array)
print('Variance of temperatures:', var_temp)

# 16. Temperatures within 20–25 degrees
count_20_25 = np.sum((temps_array >= 20) & (temps_array <= 25))
print('Temperatures within 20–25 degrees:', count_20_25)

# 17. 75th percentile
percentile_75 = np.percentile(temps_array, 75)
print('75th percentile:', percentile_75)

# 18. Temperatures greater than 26 degrees
count_gt_26 = np.sum(temps_array > 26)
print('Temperatures greater than 26 degrees:', count_gt_26)

# 19. Minimum temperature after subtracting 5 degrees
min_temp_minus_5 = np.min(temps_array - 5)
print('Minimum temperature after subtracting 5 degrees:', min_temp_minus_5)

# 20. Temperatures equal to 30.2
count_eq_30_2 = np.sum(temps_array == 30.2)
print('Temperatures equal to 30.2:', count_eq_30_2)
```

```python
sales_array = np.array([[100, 120, 90, 110], [80, 85, 95, 100], [150, 140, 130, 135]])

# 1. Number of rows and columns
num_rows, num_cols = np.shape(sales_array)
print('Number of rows:', num_rows)
print('Number of columns:', num_cols)

# 2. Average sales for the first product
avg_sales_first_product = np.mean(sales_array[0])
print('Average sales for the first product:', avg_sales_first_product)

# 3. Sales values greater than 100
count_gt_100 = np.sum(sales_array > 100)
print('Sales values greater than 100:', count_gt_100)

# 4. Total sales for the last two months
total_last_two_months = np.sum(sales_array[:, -2:], axis=0)
print('Total sales for the last two months:', total_last_two_months)

# 5. Sales values > 120 for the second product
count_gt_120_second_product = np.sum(sales_array[1] > 120)
print('Sales values > 120 for the second product:', count_gt_120_second_product)

# 6. Maximum sales value
max_sales = np.max(sales_array)
print('Maximum sales value:', max_sales)

# 7. Standard deviation for the first month
std_first_month = np.std(sales_array[:, 0])
print('Standard deviation for the first month:', std_first_month)

# 8. Sales values equal to 100
count_eq_100 = np.sum(sales_array == 100)
print('Sales values equal to 100:', count_eq_100)

# 9. Median sales for the second product
median_second_product = np.median(sales_array[1])
print('Median sales for the second product:', median_second_product)

# 10. Sales values > 100 in last two rows
count_gt_100_last_two_rows = np.sum(sales_array[1:, :] > 100)
print('Sales values > 100 in last two rows:', count_gt_100_last_two_rows)

# 11. Total sales across all products and months
total_sales = np.sum(sales_array)
print('Total sales across all products and months:', total_sales)

# 12. Sales values < 90
count_lt_90 = np.sum(sales_array < 90)
print('Sales values < 90:', count_lt_90)

# 13. Average sales for last two months
avg_last_two_months = np.mean(sales_array[:, -2:], axis=0)
print('Average sales for last two months:', avg_last_two_months)

# 14. Sales values > 110 for first product
count_gt_110_first_product = np.sum(sales_array[0] > 110)
print('Sales values > 110 for first product:', count_gt_110_first_product)
```

```python
# 1. Unique sales values in sales_array
unique_sales = np.unique(sales_array)
print('Unique sales values:', unique_sales)

# 2. Indices of sales values greater than 120 in sales_array
indices_gt_120 = np.where(sales_array > 120)
print('Indices of sales values > 120:', indices_gt_120)

# 3. Number of temps_array values greater than 25 degrees
num_temps_gt_25 = len(np.where(temps_array > 25)[0])
print('Number of temps_array values > 25 degrees:', num_temps_gt_25)

# 4. Sorted temps_array
sorted_temps = np.sort(temps_array)
print('Sorted temps_array:', sorted_temps)

# 5. Indices for sorting sales_array (flattened) in ascending order
sorted_indices_sales = np.argsort(sales_array.flatten())
print('Indices for sorting sales_array (flattened):', sorted_indices_sales)

# 6. Number of sales values in sales_array greater than 100
num_sales_gt_100 = len(np.where(sales_array > 100)[0])
print('Number of sales values > 100:', num_sales_gt_100)

# 7. Reshaped temps_array into a 5x1 array
reshaped_temps = np.reshape(temps_array, (5, 1))
print('Reshaped temps_array (5x1):', reshaped_temps)

# 8. Number of sales values in sales_array greater than 100 after adding 10 to each
num_sales_gt_100_plus_10 = np.sum((sales_array + 10) > 100)
print('Number of sales values > 100 after adding 10:', num_sales_gt_100_plus_10)

# 9. Square root of sales values in sales_array
sqrt_sales = np.sqrt(sales_array)
print('Square root of sales values:', sqrt_sales)

# 10. Number of sales values in sales_array equal to 110
num_sales_eq_110 = len(np.where(sales_array == 110)[0])
print('Number of sales values == 110:', num_sales_eq_110)

# 11. Squared sales values in sales_array
squared_sales = np.square(sales_array)
print('Squared sales values:', squared_sales)

# 12. Number of sales values in sales_array in the top 10%
top_10_percentile = np.percentile(sales_array, 90)
num_top_10_sales = np.sum(sales_array >= top_10_percentile)
print('Number of sales values in top 10%:', num_top_10_sales)

# 13. Concatenated array of temps_array and [20.0, 21.0]
concatenated_temps = np.concatenate([temps_array, np.array([20.0, 21.0])])
print('Concatenated temps_array:', concatenated_temps)

# 14. Number of sales values in sales_array ≥ 130
num_sales_ge_130 = len(np.where(sales_array >= 130)[0])
print('Number of sales values >= 130:', num_sales_ge_130)

# 15. Sum of sales values in sales_array after subtracting 5 from each
sum_sales_minus_5 = np.sum(sales_array - 5)
print('Sum of sales values after subtracting 5:', sum_sales_minus_5)
```
